<font size='4'><b>Ecart interquartile:</b>

Afin d'analyser plus pr√©cisemment nos prix, nous allons utiliser le calcul de l'√©cart interquartile:
En statistiques, l‚Äô√©cart interquartile (IQR) est une mesure de dispersion qui s'obtient en faisant la diff√©rence entre le troisi√®me et le premier quartile : IQR = Q3 - Q1

Est consid√©r√©e comme une valeure abberante une valeur inf√©rieure √† Q1‚àí1.5 IQR ou sup√©rieure √† Q3+1.5 IQR

```python
#Rep√©rage num√©rique des valeurs aberrantes:
Q1 = df_complet['price'].quantile(q=0.25, interpolation='midpoint')
Q3 = df_complet['price'].quantile(q=0.75, interpolation='midpoint')
IQR = Q3 - Q1

#Limites inf√©rieure et sup√©rieure pour qu'une valeur soit d√©finie comme une valeur aberrante:
limite_inf_outliers= Q1-1.5*IQR
limite_sup_outliers= Q3+1.5*IQR

#
outliers_inf = df_complet['price'] < limite_inf_outliers
outliers_sup = df_complet['price'] > limite_sup_outliers

print("La limite pour les valeurs aberrantes inf√©rieures est:\n", round(limite_inf_outliers,2))
print("Les index des valeurs aberrantes inf√©rieures sont:\n", np.where(outliers_inf)[0])

print("La limite pour les valeurs aberrantes sup√©rieures est:\n", round(limite_sup_outliers,2))
print("Les index des valeurs aberrantes superieures sont:\n", np.where(outliers_sup)[0])
```

```python
#Liste des vins aux prix ab√©rants :
print(df_complet.loc[(df_complet['price']>84.22),("price","post_title")])
```

La limite inf√©rieure √©tant n√©gative -27,98 aucune valeur aberrante inf√©rieure n'existe car un prix ne peut pas √™tre n√©gatif et le prix minimal est de 5.20‚Ç¨ 

La v√©rification est quand m√™me faite en affichant les index

<font size='4'><b>La m√©thode Z-score:</b>

Cette m√©thode consiste √©galement √† mettre en exergue les valeurs aberrantes

Un z-score est le nombre d‚Äô√©carts types par rapport √† la moyenne des prix. Il s‚Äôagit en fait d‚Äôune proportion du nombre d‚Äô√©carts-types en dessous ou au-dessus des prix que repr√©sente le score brut. 

Les scores Z s‚Äô√©tendent de ‚Äì 3 √©carts types (qui tomberaient √† l‚Äôextr√™me gauche du coude d‚Äôappropriation ordinaire) jusqu‚Äô√† + 3 √©carts types (qui tomberaient √† la droite la plus √©loign√©e du coude de dispersion ordinaire). 
Pour utiliser un score z, il faut conna√Ætre la moyenne Œº et en outre l‚Äô√©cart-type œÉ que nous avons pr√©c√©demment calcul√©.

Le calcul du z-score est le suivant : Z_score = (Prix - Moyenne des prix)/ IQR

```python
df_complet['Z_score']=(df_complet['price']-df_complet['price'].mean())/IQR
df_complet
```

Afin de connaitre les valeurs aberrantes, nous utilisons un seuil de 2 afin d'identifier celles-ci
Si le Z-score est sup√©rieur √† un √©cart type de 2, alors nous le d√©finissons comme une valeur aberrante

Le seuil de 2 a un indice de confiance d'environ 95%, en utilisant ce seuil nous nous apercevons que nous obtenons une liste de prix aberrants qui concerne 29 r√©f√©rences de vins ci dessous

Ce r√©sultat est tr√®s proche de la premi√®re m√©thode utilis√©e ce qui conforte notre analyse 

```python
print(df_complet.loc[(df_complet['Z_score']>2),("price","post_title","Z_score")])
```
