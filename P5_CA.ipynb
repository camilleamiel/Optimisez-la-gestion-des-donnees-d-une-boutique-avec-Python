<font size='4'><b>Ecart interquartile:</b>

Afin d'analyser plus précisemment nos prix, nous allons utiliser le calcul de l'écart interquartile:
En statistiques, l’écart interquartile (IQR) est une mesure de dispersion qui s'obtient en faisant la différence entre le troisième et le premier quartile : IQR = Q3 - Q1

Est considérée comme une valeure abberante une valeur inférieure à Q1−1.5 IQR ou supérieure à Q3+1.5 IQR

```python
#Repérage numérique des valeurs aberrantes:
Q1 = df_complet['price'].quantile(q=0.25, interpolation='midpoint')
Q3 = df_complet['price'].quantile(q=0.75, interpolation='midpoint')
IQR = Q3 - Q1

#Limites inférieure et supérieure pour qu'une valeur soit définie comme une valeur aberrante:
limite_inf_outliers= Q1-1.5*IQR
limite_sup_outliers= Q3+1.5*IQR

#
outliers_inf = df_complet['price'] < limite_inf_outliers
outliers_sup = df_complet['price'] > limite_sup_outliers

print("La limite pour les valeurs aberrantes inférieures est:\n", round(limite_inf_outliers,2))
print("Les index des valeurs aberrantes inférieures sont:\n", np.where(outliers_inf)[0])

print("La limite pour les valeurs aberrantes supérieures est:\n", round(limite_sup_outliers,2))
print("Les index des valeurs aberrantes superieures sont:\n", np.where(outliers_sup)[0])
```

```python
#Liste des vins aux prix abérants :
print(df_complet.loc[(df_complet['price']>84.22),("price","post_title")])
```

La limite inférieure étant négative -27,98 aucune valeur aberrante inférieure n'existe car un prix ne peut pas être négatif et le prix minimal est de 5.20€ 

La vérification est quand même faite en affichant les index

<font size='4'><b>La méthode Z-score:</b>

Cette méthode consiste également à mettre en exergue les valeurs aberrantes

Un z-score est le nombre d’écarts types par rapport à la moyenne des prix. Il s’agit en fait d’une proportion du nombre d’écarts-types en dessous ou au-dessus des prix que représente le score brut. 

Les scores Z s’étendent de – 3 écarts types (qui tomberaient à l’extrême gauche du coude d’appropriation ordinaire) jusqu’à + 3 écarts types (qui tomberaient à la droite la plus éloignée du coude de dispersion ordinaire). 
Pour utiliser un score z, il faut connaître la moyenne μ et en outre l’écart-type σ que nous avons précédemment calculé.

Le calcul du z-score est le suivant : Z_score = (Prix - Moyenne des prix)/ IQR

```python
df_complet['Z_score']=(df_complet['price']-df_complet['price'].mean())/IQR
df_complet
```

Afin de connaitre les valeurs aberrantes, nous utilisons un seuil de 2 afin d'identifier celles-ci
Si le Z-score est supérieur à un écart type de 2, alors nous le définissons comme une valeur aberrante

Le seuil de 2 a un indice de confiance d'environ 95%, en utilisant ce seuil nous nous apercevons que nous obtenons une liste de prix aberrants qui concerne 29 références de vins ci dessous

Ce résultat est très proche de la première méthode utilisée ce qui conforte notre analyse 

```python
print(df_complet.loc[(df_complet['Z_score']>2),("price","post_title","Z_score")])
```
